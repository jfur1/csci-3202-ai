### (1a)   Defining the Tic-Tac-Toe class structure

Fill in this class structure for Tic-Tac-Toe using what we did during class as a guide.
* `moves` is a list of tuples to represent which moves are available. Recall that we are using matrix notation for this, where the upper-left corner of the board, for example, is represented at (1,1).
* `result(self, move, state)` returns a ***hypothetical*** resulting `State` object if `move` is made when the game is in the current `state`
* `compute_utility(self, move, state)` calculates the utility of `state` that would result if `move` is made when the game is in the current `state`. This is where you want to check to see if anyone has gotten `nwin` in a row
* `game_over(self, state)` - this wasn't a method, but it should be - it's a piece of code we need to execute repeatedly and giving it a name makes clear what task the piece of code performs. Returns `True` if the game in the given `state` has reached a terminal state, and `False` otherwise.
* `utility(self, state, player)` also wasn't a method earlier, but also should be.  Returns the utility of the current state if the player is X and $-1 \cdot$ utility if the player is O.
* `display(self)` is a method to display the current game `state`, You get it for free! because this would be super frustrating without it.
* `play_game(self, player1, player2)` returns an integer that is the utility of the outcome of the game (+1 if X wins, 0 if draw, -1 if O wins). `player1` and `player2` are functional arguments that we will deal with in parts **2b** and **2d**.

Some notes:
* Assume X always goes first.
* Do **not** hard-code for 3x3 boards.
* You may add attributes and methods to these classes as needed for this problem.

```python
# Fill in as needed.
class State:
    def __init__(self, moves, symbol, board, utility):
        self.moves = moves # List[(x, y)] of possible move coordinates
        self.symbol = symbol 
        self.board = board # Dict{ (x,y) : 'X' | 'O' } represents our board
        self.utility = utility # Determines if the current state is a win

class TicTacToe:
    
    def __init__(self, nrow=3, ncol=3, nwin=3, nexp=0):
        moves = []
        for i in range(1, nrow + 1):
            for j in range(1, ncol + 1):
                moves.append( (i, j) )
        self.nrow = nrow
        self.ncol = ncol
        self.nwin = nwin
        self.nexp = nexp
        self.state = State(moves, symbol = 'X', board={}, utility=0)
        

    def result(self, move, state):
        '''
        What is the hypothetical result of move `move` in state `state` ?
        move  = (row, col) tuple where player will put their mark (X or O)
        state = a `State` object, to represent whose turn it is and form
                the basis for generating a **hypothetical** updated state
                that will result from making the given `move`
        '''
        # Invalid Move
        if move not in state.moves:
            return state       
        # Construct hypothetical game state to return
        board = state.board.copy()
        board[move] = state.symbol
        moves = list(state.moves)
        moves.remove(move)
        
        return State(moves = moves, symbol = ('O' if state.symbol == 'X' else 'X'),
                     board = board,
                     utility = self.compute_utility(move, board, state.symbol))
    
    
    def compute_utility(self, move, board, player):
        '''
        What is the utility of making move `move` in state `state`?
        If 'X' wins with this move, return 1;
        if 'O' wins return -1;
        else return 0.
        '''
        # Checks performed: [Horizontal => Vertical => Diagonal (Top-Left to Bottom-Right) =>
        #                     => Diagonal (Bottom-Left to Top-Right)]
        if(self.check_win(board, move, player, (0, 1)) or
            self.check_win(board, move, player, (1, 0)) or 
            self.check_win(board, move, player, (1, -1)) or
            self.check_win(board, move, player, (1, 1))):
            return 1 if player == 'X' else -1
        else:
            return 0

        
    def check_win(self, board, move, player, x_y_shift):
        (x_shift, y_shift) = x_y_shift
        x, y = move
        n = 0
        # Construct max line formed by move_coordinates as one endpoint
        while(board.get((x, y)) == player):
            n += 1
            x, y = x + x_shift, y + y_shift
        x, y = move  # Reset move and perform line test on other side of the point
        while(board.get((x, y)) == player):
            n += 1
            x, y = x - x_shift, y - y_shift
        n -= 1  # Subtract one since we double checked move coordinate
        return n >= self.nwin  # Returns the boolean value of the inequality

    
    def game_over(self, state):
        '''game is over if someone has won (utility!=0) or there
        are no more moves left'''
        # Returns the boolean value of the expression
        return state.utility != 0 or len(state.moves) == 0
        
    def utility(self, state, player):
        '''Return the value to player; 1 for win, -1 for loss, 0 otherwise.'''
        return state.utility if player == 'X' else -state.utility

    def display(self, state):
        for row in range(1, self.nrow+1):
            for col in range(1, self.ncol+1):
                print(state.board.get((row, col), '.'), end=' ')
            print()
        
    def play_game(self, player1=0, player2=0, return_nexp = False, prune = True):
        '''Play a game of tic-tac-toe!'''
        state = self.state
        nexp = 0
        
        while(True):
            for player in range(1, 3):

                move = None
                if(not self.game_over(state)):
                    if state.moves:
                        # Two Random CPU Players
                        if(player1 == 1 and player2 == 1):
                            move = random_player(state)
                            
                        # Alpha Beta player vs. Random CPU
                        elif(player1 == 2 or player2 == 2) and (player1 == 1 or player2 == 1):
                            if(player == 1):
                                (move, nexp) = alphabeta_player(self, state, nexp, prune)
                            else:
                                move = random_player(state)
                                
                        # Alpha Beta vs. Alpha Beta
                        elif(player1 == 2) and (player2 == 2):
                            (move, nexp) = alphabeta_player(self, state, nexp, prune)
                            
                        # One Random Player -- Human goes first
                        elif(player1 == 1 or player2 == 1) and (player != 1):
                            move = random_player(state)
                            print("CPU chooses location: ", move)
                            print("")
                            
                        # Two humans
                        else:
                            print("Current State: ")
                            self.display(state)
                            print("Player: {}".format(state.symbol))
                            print("Available moves: {}".format(state.moves))
                            
                            move_string = input('Your move: ')
                            print("")
                            try:
                                move = eval(move_string)
                            except NameError:
                                move = move_string
                    else:
                        print('no legal moves: passing turn to next player')
                        continue
                state = self.result(move, state)
                
                if(self.game_over(state)):
                    if return_nexp:
                        if prune:
                            print("Number of states expanded by minimax with pruning:", nexp)
                        else:
                            print("Number of states expanded by minimax without pruning:", nexp)
                    if not(player1 == 0 and player2 == 0):
                        return state.utility
                    print("Current State: ")
                    self.display(state)
                    symbol = ('O' if state.symbol == 'X' else 'X')
                    print("")
                    print(symbol, "wins!")
                    return state.utility
```

### (1d) Define an alpha-beta player

Alright. Let's finally get serious about our Tic-Tac-Toe game.  No more fooling around!

Craft a function called `alphabeta_player` that takes a single argument of a `TicTacToe` class object and returns the minimax move in the `state` of the `TicTacToe` game. As the name implies, this player should be implementing alpha-beta pruning as described in the textbook and lecture.

Note that your alpha-beta search for the minimax move should include function definitions for `max_value` and `min_value` (see the aggressively realistic pseudocode from the lecture slides).

In your code for the `play_game` method above, make sure that `alphabeta_player` could be a viable input for the `player1` and/or `player2` arguments.

```python
def alphabeta_player(TicTacToe, state, nexp, prune):
    player = state.symbol
    game = TicTacToe
    
    def max_value(state, alpha, beta, nexp):
        if game.game_over(state):
            return (game.utility(state, player), nexp)
        v = -np.inf
        for a in state.moves:
            (v_min, nexp) = min_value(game.result(a, state), alpha, beta, nexp)
            v = max(v, v_min)
            if (v >= beta and prune):
                nexp += 1
                return (v, nexp)
            alpha = max(v, alpha)
        nexp += 1
        return (v, nexp)
        
    def min_value(state, alpha, beta, nexp):
        if game.game_over(state):
            return (game.utility(state, player), nexp)
        v = np.inf
        for a in state.moves:
            (v_max, nexp) = max_value(game.result(a, state), alpha, beta, nexp)
            v = min(v, v_max)
            if (v <= alpha and prune):
                nexp += 1
                return (v, nexp)
            beta = min(beta, v)
        nexp += 1
        return (v, nexp)
        
    alpha = -np.inf
    beta = np.inf
    best_action = None
    
    for a in state.moves:
        (v, nexp) = min_value(game.result(a, state), alpha, beta, nexp)
        if v > alpha:
            alpha = v
            best_action = a
            
    return (best_action, nexp)
```

```python
class Individual(object):
    
    def __init__(self, chromosome):
        self.chromosome = chromosome
        self.fitness = self.calc_fitness
        
    def mutate(self, p, i):
        prob = np.random.randint(1, 101)
        gene = self.chromosome[i]
        if prob >= (p*10):
            gene = round(random.uniform(0.5, 108.5), 1)
        return gene
    
    def create(self, p):
        global target
        genome_len = len(target)-1
        return[self.mutate(p, i) for i in range(0, genome_len)]
        
    def mate(self, par2, p):
        child_chromosome = []
        for gp1, gp2 in zip(self.chromosome, par2.chromosome):
        
            prob = np.random.randint(0, 101)
        
            if prob < 45:
                child_chromosome.append(gp1)
                
            elif prob < 90:
                child_chromosome.append(gp2)
                
            else:
                child_chromosome.append(self.mutate(p, len(child_chromosome)))
            
        return Individual(child_chromosome)
        
    def calc_fitness(self):
        fitness = 0
        global target
        for i in range(len(self.chromosome)):
            #self.chromosome[i] = '{0:g}'.format(self.chromosome[i])
            if self.chromosome[i] != target[i]:
                fitness += 1

        return fitness
```

```python
def generate(target, iterations = 200, population_size = 10, p = 0.1):
    generation = 1
    i = 0
    found = False
    population = []
    
    for _ in range(population_size):
        obj = Individual([i for i in range(0, len(target))])
        genome = Individual.create(obj, p)
        population.append(Individual(genome))
        
    while not found:
        i += 1
        population = sorted(population, key = lambda x:x.fitness())
        
        if population[0].fitness() <= 0:
            found = True
            print("Target Recipe Acquired!")
            return
        
        new_population = []
        
        s = int((10*population_size)/100)
        new_population.extend(population[:s])
        
        s = int((90*population_size)/100)
        for _ in range(s):
            parent1 = np.random.choice(population[:50])
            parent2 = np.random.choice(population[:50])
            child = parent1.mate(parent2, p)
            new_population.append(child)
            
        population = new_population
        
        value = "".join(str(population[0].chromosome))

        print("Generation: {}\tGenome: {}\tFitness: {}".format(generation, value ,population[0].fitness())) 
        
        generation += 1
        if i >= iterations:
            print("Max. Number of Iterations Reached")
            return
    value = "".join(str(population[0].chromosome))
    print("Generation: {}\tGenome: {}\tFitness: {}".format(generation, value ,population[0].fitness())) 
```

```python
generate(target)
```

**By using a population size of 10 for 200 generations, our genetic algorithm does not successfully find the target recipe; however, it is making progress as the fitness converges on the correct recipe.
